```python
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
import random
from sklearn.model_selection import train_test_split

# Load the image

DATADIR = "./test_data/"
SAVEDIR = "./new_data/"
CATEGORIES = ["person", "bike", "car"]
```

```python
string = ['thermal_00002.jpeg', 3, 396, 200, 543, 250]

image_file = [string[0]]
class_id = [int(string[1])-1]
x_min = [int(string[2])]
x_max = [int(string[3])]
y_min = [int(string[4])]
y_max = [int(string[5])]
```

```python
for index in range(len(image_file)):
    # load image
    path = os.path.join(DATADIR, image_file[index])
    img_array = cv2.imread(path, cv2.IMREAD_GRAYSCALE)

    # Check if image loaded successfully
    if img_array is None:
        raise FileNotFoundError(f"Error: Image '{image_file}' not found at '{DATADIR}'.")

    # crop image
    cropped_image = img_array[y_max[index]:y_min[index], x_max[index]:x_min[index]]

    #save into respective category folder , then show image
    save_path = os.path.join(SAVEDIR, CATEGORIES[class_id[index]])
    print(save_path)
    os.makedirs(save_path, exist_ok=True)

    cv2.imwrite(save_path+f'/image_{index}.jpeg', cropped_image)


    print(img_array.shape)
    print(cropped_image.shape)
    plt.imshow(img_array, cmap="gray")
    plt.show()
    plt.imshow(cropped_image, cmap="gray")
    plt.show()
```
